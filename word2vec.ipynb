```python
All_posts_word = ""
wordlist = []
for i in shared_task_posts["post_body"]:
    i = '%s' %i
    i = i.upper()
    wordlist.append(i)

# wordlist
All_posts_word = " ".join(wordlist)
fh = open('All_posts_word.txt', 'w', encoding='utf-8')
fh.write(All_posts_word)
fh.close()
# subreddit_word
```

```python
def model_train(train_file_name, save_model_file):  # model_file_name,save_model

    # 
    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
    sentences = word2vec.Text8Corpus(train_file_name)  # 
    model = gensim.models.Word2Vec(sentences, size=200)  # skip-gram; window=5
    model.save(save_model_file)
    model.wv.save_word2vec_format(save_model_file + ".bin", binary=True)   # 
    
model_train("C:\\Users\\zxz\\Desktop\\code\\clp2019\\All_posts_word.txt","D:\\workplace\\word2vec.model" )
```
